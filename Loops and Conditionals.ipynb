```python
secret = 7
```

```python
guess = 5
```

```python
if guess < secret:
    print("too low")
elif guess > secret:
    print ("too high")
else:
    print("just right")
```

```python
small = True
```

```python
green = True
```

```python
if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```
